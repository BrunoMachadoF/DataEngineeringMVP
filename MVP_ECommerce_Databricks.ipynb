## Pipeline

Todo o trabalho de ETL será realizado dentro do Databricks seguindo os passos descritos abaixo:

1. **Extração:**
   - Os dados serão extraídos da plataforma Kaggle, especificamente do dataset "Brazilian E-Commerce Public Dataset by Olist". Este dataset contém informações detalhadas sobre pedidos, produtos, vendedores, clientes, métodos de pagamento, avaliações e geolocalização.

2. **Transformação:**
   - Dentro do Databricks, os dados extraídos serão limpos e transformados para garantir consistência e qualidade. Isso inclui o tratamento de valores ausentes, a padronização de formatos, a tradução de categorias de produtos e a criação de novas colunas derivadas para análises mais detalhadas.

3. **Carregamento:**
   - Após a transformação, os dados serão carregados em tabelas otimizadas no Databricks, prontas para análise. O armazenamento será feito utilizando o Amazon S3, aproveitando os 15 GB de armazenamento disponíveis na Databricks Community Edition.

4. **Análise:**
   - Com os dados carregados, serão realizadas diversas análises exploratórias e descritivas para responder às perguntas de negócio formuladas.

Esta abordagem garante que todo o processo de ETL (Extração, Transformação e Carregamento) seja eficiente e integrado, permitindo uma análise robusta e escalável dos dados de comércio eletrônico brasileiro.

### Esquema
A especificação do esquema de um Data Warehouse utiliza o paradigma multidimensional como representação conceitual. Toda modelagem dimensional, por sua vez, possui dois elementos imprescindíveis: Fatos e Dimensões. Os fatos representam as medidas de interesse, quase sempre numéricas, enquanto as dimensões representam as perspectivas do negócio, sobre as quais se deseja analisar essas medidas de interesse.

Assim, será utilizado um esquema multidimensional do tipo floco de neve para a modelagem de dados, onde a tabela fato corresponde às vendas, mais especificamente aos itens vendidos. Cada venda está associada a três chaves estrangeiras que referenciam as chaves primárias das tabelas dimensão Vendedor, Produto e Pedido. Os fatos representados pela tabela fato são a quantidade vendida do item, seu valor e o valor de frete associado. Por fim, as dimensões Cliente, Pagamento, Review e Tempo estão ligadas à tabela dimensão Pedido, sendo essa uma característica de um esquema floco de neve.

**7.1 Tratamento dos valores nulos**

Vamos desenvolver um tratamento adequado para cada uma das colunas da tabela products que contêm valores nulos, com explicações para os tratamentos aplicados.

Remover registros com product_id nulo: product_id é uma chave primária, portanto, registros sem esse valor não são válidos e devem ser removidos.
Tratar valores nulos em product_category_name: Substituir valores nulos por uma categoria genérica, como "unknown", para manter a integridade do registro.
Tratar valores nulos em product_name_lenght e product_description_lenght: Substituir valores nulos pela média das respectivas colunas.
Tratar valores nulos em product_photos_qty: Substituir valores nulos pelo valor zero, assumindo que produtos sem valor informado não possuem fotos.
Tratar valores nulos em product_weight_g, product_length_cm, product_height_cm e product_width_cm: Substituir valores nulos pela média das respectivas colunas.
Aqui está o código de tratamento:

```python
from pyspark.sql.functions import col, when, mean, lit

# 1. Remover registros com product_id nulo
products_df = products_df.filter(col("product_id").isNotNull())

# 2. Tratar valores nulos em product_category_name
products_df = products_df.withColumn("product_category_name",
                                     when(col("product_category_name").isNotNull(), col("product_category_name"))
                                     .otherwise(lit("unknown")))

# 3. Tratar valores nulos em product_name_lenght e product_description_lenght
mean_name_length = products_df.select(mean(col("product_name_lenght"))).first()[0]
mean_description_length = products_df.select(mean(col("product_description_lenght"))).first()[0]

products_df = products_df.withColumn("product_name_lenght",
                                     when(col("product_name_lenght").isNotNull(), col("product_name_lenght"))
                                     .otherwise(lit(mean_name_length).cast("int")))

products_df = products_df.withColumn("product_description_lenght",
                                     when(col("product_description_lenght").isNotNull(), col("product_description_lenght"))
                                     .otherwise(lit(mean_description_length).cast("int")))

# 4. Tratar valores nulos em product_photos_qty
products_df = products_df.withColumn("product_photos_qty",
                                     when(col("product_photos_qty").isNotNull(), col("product_photos_qty"))
                                     .otherwise(lit(0)))

# 5. Tratar valores nulos em product_weight_g, product_length_cm, product_height_cm e product_width_cm
mean_weight = products_df.select(mean(col("product_weight_g"))).first()[0]
mean_length = products_df.select(mean(col("product_length_cm"))).first()[0]
mean_height = products_df.select(mean(col("product_height_cm"))).first()[0]
mean_width = products_df.select(mean(col("product_width_cm"))).first()[0]

products_df = products_df.withColumn("product_weight_g",
                                     when(col("product_weight_g").isNotNull(), col("product_weight_g"))
                                     .otherwise(lit(mean_weight).cast("int")))

products_df = products_df.withColumn("product_length_cm",
                                     when(col("product_length_cm").isNotNull(), col("product_length_cm"))
                                     .otherwise(lit(mean_length).cast("int")))

products_df = products_df.withColumn("product_height_cm",
                                     when(col("product_height_cm").isNotNull(), col("product_height_cm"))
                                     .otherwise(lit(mean_height).cast("int")))

products_df = products_df.withColumn("product_width_cm",
                                     when(col("product_width_cm").isNotNull(), col("product_width_cm"))
                                     .otherwise(lit(mean_width).cast("int")))

# Exibir o esquema final e a contagem de registros
products_df.printSchema()
print("Total de registros após tratamento:", products_df.count())
```

```python
from pyspark.sql.functions import round, avg, datediff

# Calculando o tempo de entrega
delivery_times = orders_df.withColumn("delivery_time", datediff("order_delivered_customer_date", "order_purchase_timestamp"))

# Calculando o tempo médio de entrega por região
avg_delivery_time_by_region = delivery_times.join(customers_geo_df, "customer_id") \
    .groupBy("geolocation_state") \
    .agg(round(avg("delivery_time"), 2).alias("avg_delivery_time")) \
    .orderBy(col("avg_delivery_time").desc())

# Exibindo o tempo médio de entrega por região
avg_delivery_time_by_region.show()
```

```python
# Obter os dados em Pandas DataFrame
avg_delivery_time_by_region_pd = avg_delivery_time_by_region.toPandas()

# Plotar o gráfico do tempo médio de entrega
plt.figure(figsize=(12, 6))
sns.barplot(data=avg_delivery_time_by_region_pd, x="geolocation_state", y="avg_delivery_time")
plt.title("Tempo Médio de Entrega por Região")
plt.xlabel("Estado")
plt.ylabel("Tempo Médio de Entrega (dias)")
plt.xticks(rotation=45)
plt.show()
```

Para resolver a questão "Qual é o tempo médio de entrega e como ele varia entre diferentes regiões?", foi realizada uma análise detalhada utilizando PySpark. Primeiramente, foi calculado o tempo de entrega de cada pedido subtraindo a data de compra da data de entrega ao cliente. Em seguida, foi calculado o tempo médio de entrega por região (estado) agrupando os dados por estado e calculando a média dos tempos de entrega.

Os resultados mostraram o tempo médio de entrega em dias para cada estado. Posteriormente, esses dados foram convertidos para um DataFrame do Pandas e um gráfico de barras foi gerado para visualizar a distribuição do tempo médio de entrega por região. O gráfico permitiu identificar facilmente quais estados têm os maiores e menores tempos médios de entrega, proporcionando uma visão clara das variações regionais.

**5. Quais categorias de produtos têm os maiores índices de satisfação do cliente?**

**6. Existe correlação entre preço e taxa de satisfação do cliente?**

```python
# Calculando a correlação entre preço e taxa de satisfação
price_satisfaction_correlation = reviews_products_df \
    .select(corr("price", "review_score").alias("correlation_price_review_score"))

# Exibindo a correlação
price_satisfaction_correlation.show()
```

Para investigar a existência de correlação entre o preço e a taxa de satisfação do cliente, foi realizada uma análise de correlação utilizando as tabelas de avaliações e produtos. A correlação foi calculada entre as colunas de preço dos produtos e a pontuação das avaliações dadas pelos clientes.

Os resultados indicam que a correlação entre o preço e a taxa de satisfação do cliente é muito baixa, com um valor de correlação de aproximadamente -0.083. Este valor sugere que não há uma relação linear significativa entre o preço dos produtos e a satisfação dos clientes. Em outras palavras, o preço dos produtos não parece influenciar diretamente a satisfação do cliente de forma relevante.

Essa informação é importante para estratégias de precificação e marketing, pois indica que fatores além do preço podem estar desempenhando um papel mais significativo na determinação da satisfação do cliente.
